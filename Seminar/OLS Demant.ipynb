```python
import pandas as pd
import statsmodels.api as sm
```

```python
Nord = pd.read_csv('SP.csv', parse_dates=True, index_col='Date',)
Nord
```

```python
Nord1 = Nord[Nord.columns[Nord.columns.isin(['Date','Last Price', 'Last Price SP'])]]
Nord1['Last Price'] = Nord1['Last Price'].str.replace('.','', regex=True).str.replace(',', '.', regex=True).astype(float)
Nord1['Last Price SP'] = Nord1['Last Price SP'].str.replace('.','', regex=True).str.replace(',', '.', regex=True).astype(float)
Nord1 = Nord1.iloc[1:, :]
Nord1
```

```python
Nord1.sort_values(by='Date', inplace = True) 
Nord1
```

```python
daily_ret = Nord1.pct_change(1)
daily_ret_clean = daily_ret.dropna(axis=0)
daily_ret_clean
```

```python

# split dependent and independent variable
X = daily_ret_clean['Last Price SP']
y = daily_ret_clean['Last Price']

# Add a constant to the independent value
X1 = sm.add_constant(X)

# make regression model 
model = sm.OLS(y, X1)

# fit model and print results
results = model.fit()
print(results.summary())
```

```python
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(X, y)

print(f'Beta = {slope}')
```
